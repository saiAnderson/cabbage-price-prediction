# 把用不到的 row 刪掉 (因為有些是 useless 資料)

```python
import csv

def delete_rows_csv(file_path):
    # find these row (want to delete)
    rows_to_delete = [0,1,2,3] 

    # Get the total number of rows in the CSV file 
    with open(file_path, 'r', newline='',encoding="utf-8") as file:
        reader = csv.reader(file)
        total_rows = sum(1 for row in reader) #in order to delete the last row

    # Add the index of the last row to the rows_to_delete list
    rows_to_delete.append(total_rows - 1)

    # Read the CSV file and filter out rows to delete
    with open(file_path, 'r', newline='',encoding="utf-8") as file:
        reader = csv.reader(file)
        lines = [line for i, line in enumerate(reader) if i not in rows_to_delete]

    # Write the remaining lines back to the CSV file
    with open(file_path, 'w', newline='',encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerows(lines)


file_path1 = r'2021_高麗菜.csv'
file_path2 = r'2022_高麗菜.csv'
delete_rows_csv(file_path1)
delete_rows_csv(file_path2)
```

```python
df1 = pd.read_csv(r'2021_高麗菜.csv')
df2 = pd.read_csv(r'2022_高麗菜.csv')

print(df1.columns)
print(df2.columns)
```

# 把 useless column 刪掉 和 處理 column 名稱(空白問題)

```python
def fix_columns(df):
    # drop useless columns
    df.drop(labels=['3','Unnamed: 11'],axis='columns',inplace=True)

    new_column_names = {'日　　期': '日期', '市　　場': '市場','產　　品':'產品'}
    df.rename(columns=new_column_names, inplace=True)
    return df

df1 = fix_columns(df1)
df2 = fix_columns(df2)

df1.columns
```

```python
def to_year(row):
    y_string = row['日期']
    y_string = y_string[:y_string.find('/')]
    y = int(y_string)+1911
    return y

def to_month(row):
    y_string = row['日期']
    y_string = y_string[y_string.find('/')+1:y_string.rfind('/')]
    return int(y_string)

def to_day(row):
    y_string = row['日期']
    y_string = y_string[y_string.rfind('/')+1:]
    return int(y_string)

df1['year'] = df1.apply(to_year, axis=1)
df1['month'] = df1.apply(to_month, axis=1)
df1['day'] = df1.apply(to_day, axis=1)

df2['year'] = df2.apply(to_year, axis=1)
df2['month'] = df2.apply(to_month, axis=1)
df2['day'] = df2.apply(to_day, axis=1)
```

```python
from datetime import datetime
def to_day_of_yeaer(row):
    specified_date = datetime(row['year'], row['month'], row['day'])
    day_of_yearr = specified_date.timetuple().tm_yday
    return day_of_yearr

df1['day_of_year'] = df1.apply(to_day_of_yeaer, axis=1)
df2['day_of_year'] = df2.apply(to_day_of_yeaer, axis=1)
```

```python
df1.to_csv('Train_2021_高麗菜_data_file.csv', index=False)
df2.to_csv('Train_2022_高麗菜_data_file.csv', index=False)
```
