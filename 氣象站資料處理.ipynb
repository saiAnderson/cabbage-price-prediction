```python
# 得到日期
def get_date_from_filename(dirpath):
    csv_file_names_without_csv = []

    for i in dirpath:
        csv_file_names_without_csv.append(i[i.find('-20')+1:i.find('.')])
    return csv_file_names_without_csv

year_month_2021 = get_date_from_filename(file_path_2021)
year_month_2022 = get_date_from_filename(file_path_2022)
```

```python
# 先把處理好的資料(dataframe)放到 list 中，最後在合併
def combine_all_csv_file(file_path,year_month):
    dfs = []

    for i in range(len(file_path)):
        df = pd.read_csv(file_path[i])
        # 刪掉第一個 row 
        df = df.drop(df.index[0])
        df['觀測站'] = '宜蘭縣'
        df['年份'] =  int(year_month[i][:year_month[i].find('-')])
        df['月'] = int(year_month[i][year_month[i].find('-')+1:])

        # Append the modified DataFrame to the list
        dfs.append(df)

    # Concatenate the list of DataFrames into a single DataFrame
    final_df = pd.concat(dfs, ignore_index=True)

    return final_df

final_weather_file_2021 = combine_all_csv_file(file_path_2021,year_month_2021)
final_weather_file_2022 = combine_all_csv_file(file_path_2022,year_month_2022)
```

# using all function 
you just input the "year" in "complete" function it will generate that year's weather information in one csv , but this only "宜蘭-觀測站"

```python
def complete(year):
    dirpath = str(year)+"-宜蘭-觀測站"
    file_path = complete_path(dirpath)
    year_month = get_date_from_filename(file_path)
    final_weather_file = combine_all_csv_file(file_path,year_month)
    # final_weather_file['day_of_year'] = final_weather_file.apply(to_day_of_yeaer, axis=1)
    file_name = "all_"+ str(year) + "宜蘭_weather_file.csv"
    # change colums's name
    new_columns = {'觀測時間(day)': 'day'}
    final_weather_file = final_weather_file.rename(columns=new_columns)
    final_weather_file['day'] = final_weather_file['day'].astype(int)
    return final_weather_file
```

# combine all years

```python
years = [2018,2019,2020,2021,2022]
dfs = []
for i in years:
    dfs.append(complete(i))

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.to_csv('all_weather_data.csv', index=False)
```

```python
def complete_path(dirpath):
    paths = []
    for f in os.listdir(dirpath):
        fullpath = os.path.join(dirpath,f)
        paths.append(fullpath)
    return paths

def get_date_from_filename(dirpath):
    csv_file_names_without_csv = []

    for i in dirpath:
        csv_file_names_without_csv.append(i[i.find('-20')+1:i.find('.')])
    return csv_file_names_without_csv

def web_crawler_combine_all_csv_file(file_path,year_month):
    dfs = []

    for i in range(len(file_path)):
        df = pd.read_csv(file_path[i])
        # 刪掉第一個 row 
        # Drop the first row
        df = df.drop(1)

        # Set the second row as the new header
        df.columns = df.iloc[0]
        df = df[1:]
        df.columns = df.columns.str.replace(' ', '')

        df['觀測站'] = '宜蘭縣'
        df['year'] =  int(year_month[i][:year_month[i].find('-')])
        df['month'] = int(year_month[i][year_month[i].find('-')+1:])

        # Append the modified DataFrame to the list
        dfs.append(df)

    # Concatenate the list of DataFrames into a single DataFrame
    final_df = pd.concat(dfs, ignore_index=True)
    return final_df

def hand_get_combine_all_csv_file(file_path,year_month):
    dfs = []

    for i in range(len(file_path)):
        df = pd.read_csv(file_path[i])
        # 刪掉第一個 row 
        df = df.drop(df.index[0])
        df['觀測站'] = '宜蘭縣'
        df['year'] =  int(year_month[i][:year_month[i].find('-')])
        df['month'] = int(year_month[i][year_month[i].find('-')+1:])

        # Append the modified DataFrame to the list
        dfs.append(df)

    # Concatenate the list of DataFrames into a single DataFrame
    final_df = pd.concat(dfs, ignore_index=True)
    return final_df


# def to_day_of_yeaer(row):
#     specified_date = datetime(row['year'], row['month'], int(row['觀測時間(day)']))
#     day_of_yearr = specified_date.timetuple().tm_yday
#     return day_of_yearr
```

```python
def complete(year):
    dirpath = str(year)+"-宜蘭-觀測站"
    file_path = complete_path(dirpath)
    year_month = get_date_from_filename(file_path)

    if(year==2017): #because using when using the web crawler the result may have some different so we have to separate
        final_weather_file = web_crawler_combine_all_csv_file(file_path,year_month)
    else:
        final_weather_file = hand_get_combine_all_csv_file(file_path,year_month)

    # final_weather_file['day_of_year'] = final_weather_file.apply(to_day_of_yeaer, axis=1)
    file_name = "all_"+ str(year) + "宜蘭_weather_file.csv"
    # change colums's name
    new_columns = {'觀測時間(day)': 'day'}
    final_weather_file = final_weather_file.rename(columns=new_columns)
    final_weather_file['day'] = final_weather_file['day'].astype(int)
    return final_weather_file
```

```python
years = [2017,2018,2019,2020,2021,2022]
dfs = []
for i in years:
    dfs.append(complete(i))

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.to_csv('all_weather_data.csv', index=False)
```
