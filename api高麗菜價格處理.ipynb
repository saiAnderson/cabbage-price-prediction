```python
import pandas as pd
import numpy as np
import os
import csv
from datetime import datetime
```

```python
df = pd.read_csv(r'2016-2022_宜蘭市果菜市場價格.csv')
df.head(3)
```

```python
df.drop(labels=['Unnamed: 0'],axis='columns',inplace=True)
df.head(3)
```

```python
def to_year(row):
    y_string = row['TransDate']
    y_string = y_string[:y_string.find('.')]
    y = int(y_string)+1911
    return y

def to_month(row):
    y_string = row['TransDate']
    y_string = y_string[y_string.find('.')+1:y_string.rfind('.')]
    return int(y_string)

def to_day(row):
    y_string = row['TransDate']
    y_string = y_string[y_string.rfind('.')+1:]
    return int(y_string)

def to_day_of_yeaer(row):
    specified_date = datetime(row['year'], row['month'], row['day'])
    day_of_yearr = specified_date.timetuple().tm_yday
    return day_of_yearr
```

# 在 dataset 裡面創造新的 columns "year" "month" "day"

```python
df['year'] = df.apply(to_year, axis=1)
df['month'] = df.apply(to_month, axis=1)
df['day'] = df.apply(to_day, axis=1)

df.head(4)
```
