T 表示微量(小於 0.5mm)，x 表故障，& 代表降水量資料累積於後，V 表風向不定，/表不
明

```python
def fix_rainfall_data(df):
    df['降水量(mm)'] = df['降水量(mm)'].replace('T', '0.0')
    df['降水量(mm)'] = df['降水量(mm)'].replace('X', '-1')
    df['降水量(mm)'] = pd.to_numeric(df['降水量(mm)'])
    return df

fix_rainfall_data(df_weather)['降水量(mm)']
```

```python
def Max_10_mintes_fix_rainfall_data(df):
    df['最大十分鐘降水量(mm)'] = df['最大十分鐘降水量(mm)'].replace('T', '0.0')
    df['最大十分鐘降水量(mm)'] = df['最大十分鐘降水量(mm)'].replace('X', '-1')
    df['最大十分鐘降水量(mm)'] = pd.to_numeric(df['最大十分鐘降水量(mm)'])
    return df

def Max_60_mintes_fix_rainfall_data(df):
    df['最大六十分鐘降水量(mm)'] = df['最大六十分鐘降水量(mm)'].replace('T', '0.0')
    df['最大六十分鐘降水量(mm)'] = df['最大六十分鐘降水量(mm)'].replace('X', '-1')
    df['最大六十分鐘降水量(mm)'] = pd.to_numeric(df['最大六十分鐘降水量(mm)'])
    return df

def A_type_evaporation(df):
    df['A型蒸發量(mm)'] = df['A型蒸發量(mm)'].replace('/', '-1')
    df['A型蒸發量(mm)'] = pd.to_numeric(df['A型蒸發量(mm)'])
    return df

def All_sky_insolation(df):
    df['全天空日射量(MJ/㎡)'] = df['全天空日射量(MJ/㎡)'].replace('/', '-1')
    df['全天空日射量(MJ/㎡)'] = df['全天空日射量(MJ/㎡)'].replace('X', '-1')
    df['全天空日射量(MJ/㎡)'] = df['全天空日射量(MJ/㎡)'].replace('--', '-1')
    df['全天空日射量(MJ/㎡)'] = pd.to_numeric(df['全天空日射量(MJ/㎡)'])
    return df

def highest_UV_index_of_the_day(df):
    df['日最高紫外線指數'] = df['日最高紫外線指數'].replace('/', '-1')
    df['日最高紫外線指數'] = df['日最高紫外線指數'].replace('X', '-1')
    df['日最高紫外線指數'] = df['日最高紫外線指數'].replace('--', '-1')
    df['日最高紫外線指數'] = pd.to_numeric(df['日最高紫外線指數'])
    return df


Max_10_mintes_fix_rainfall_data(df_weather)
Max_60_mintes_fix_rainfall_data(df_weather)
A_type_evaporation(df_weather)
All_sky_insolation(df_weather)
highest_UV_index_of_the_day(df_weather)
```

```python
#時間軸反轉
merged.sort_values('TransDate', ascending=True, inplace=True)

#交易量上移一格
merged['Trans_Quantity'] = merged['Trans_Quantity'].shift(-1)
merged['Trans_Quantity'].fillna(0, inplace=True)
```

```python
# Define the columns for which you want to calculate the average
columns_to_average = ['測站氣壓(hPa)', '海平面氣壓(hPa)','測站最高氣壓(hPa)','測站最低氣壓(hPa)', '氣溫(℃)'
                      , '最高氣溫(℃)', '最低氣溫(℃)','露點溫度(℃)' ,'相對溼度(%)','最小相對溼度(%)','最大瞬間風(m/s)',
                      '降水量(mm)','降水時數(hour)','最大十分鐘降水量(mm)','最大六十分鐘降水量(mm)','日照時數(hour)','日照率(%)',
                      '能見度(km)','A型蒸發量(mm)','總雲量(0~10)','地溫0cm','地溫10cm','地溫20cm','地溫30cm','地溫50cm','地溫100cm']

# Calculate the average for each column based on the previous day's values
for column in columns_to_average:
    merged[column + '_avg60'] = merged[column].shift(1).rolling(window=60).mean()

for column in columns_to_average:
    merged[column + '_avg30'] = merged[column].shift(1).rolling(window=30).mean()

for column in columns_to_average:
    merged[column + '_avg15'] = merged[column].shift(1).rolling(window=15).mean()

# Drop rows with NaN values (first row, as there is no previous day)
merged = merged.dropna()
```

前一天的 middle price and Avg_Price

```python
merged['last_day_Middle_Price'] = merged['Middle_Price'].shift(1)
merged['last_day_Avg_Price'] = merged['Avg_Price'].shift(1)
merged = merged.dropna()
```
